```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
fr = np.load('final_results.npy')
```

```python
fr
```

```python
TTCs = fr[:, 0]
RMSE = fr[:, 1]
MaxE = fr[:, 2]
NumG = fr[:, 3]
num_universes = fr.shape[0]
seeds = np.array([1218399353,  261779297, 1830636481,  717930975,  925300214,  800619223,
  258378443,  788299983, 1576600005,  531766210,  829272384, 1075557653,
  1282776693, 1351570620,  870490860,  947266021,  981106429, 1466557164,
  1160698887, 2061977524])
```

```python
# plot TTC over universes
plt.plot(np.arange(num_universes), TTCs, c='orange')
plt.scatter(np.argsort(TTCs)[:3], TTCs[np.argsort(TTCs)[:3]], c='r')
plt.xlabel('Universes')
plt.ylabel('Time to converge')
plt.title('Time to converge for multiple universes (seconds)')
plt.xticks(np.arange(num_universes), np.arange(num_universes) + 1)
plt.text(s='The best seeds were: {}'.format(seeds[np.argsort(TTCs)[:3]]), x=0, y=150, fontsize=8, color='green')
# plt.legend()
# plt.savefig('./outputs/TTC_Universes_2')
plt.show()
plt.close()
```

```python
# plot RMS error over universes
plt.plot(np.arange(num_universes), RMSE, c='orange')
plt.scatter(np.argsort(RMSE)[:3], RMSE[np.argsort(RMSE)[:3]], c='r')
plt.xlabel('Universes')
plt.ylabel('RMS Error')
plt.title('RMS Error for multiple universes')
plt.xticks(np.arange(num_universes), np.arange(num_universes) + 1)
plt.text(s='The best seeds were: {}'.format(seeds[np.argsort(RMSE)[:3]]), x=0, y=0.14, fontsize=8, color='green')
# plt.savefig('./outputs/RMSE_Universes_2')
plt.show()
plt.close()
```

```python
 # plot max error over universes
plt.plot(np.arange(num_universes), MaxE, c='orange')
plt.scatter(np.argsort(MaxE)[:3], MaxE[np.argsort(MaxE)[:3]], c='r')
plt.xlabel('Universes')
plt.ylabel('Max Error')
plt.title('Max Error for multiple universes')
plt.xticks(np.arange(num_universes), np.arange(num_universes) + 1)
plt.text(s='The best seeds were: {}'.format(seeds[np.argsort(MaxE)[:3]]), x=0, y=0.41, fontsize=8, color='green')
# plt.savefig('./outputs/MaxE_Universes_2')
plt.show()
plt.close()
```

```python
# plot number of generations to converge over universes
plt.plot(np.arange(num_universes), NumG, c='orange')
plt.scatter(np.argsort(NumG)[:3], NumG[np.argsort(NumG)[:3]], c='r')
plt.xlabel('Universes')
plt.ylabel('Number of generations till convergence')
plt.title('Generations till convergence for multiple universes')
plt.xticks(np.arange(num_universes), np.arange(num_universes) + 1)
plt.text(s='The best seeds were: {}'.format(seeds[np.argsort(NumG)[:3]]), x=4, y=12, fontsize=7, color='green')
# plt.show()
# plt.savefig('./outputs/GenC_Universes_2')
plt.show()
plt.close()
```
